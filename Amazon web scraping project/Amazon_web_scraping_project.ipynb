# Tryng to scrape more data

```python

# Find all product containers
items = soup.find_all("div", attrs={'data-component-type': 's-search-result'})

all_products = []

for item in items:
    # We extract data from EACH item container
    try:
        # Title: Usually inside an <h2> tag in search results
        title = item.find("h2").get_text(strip=True)

        # Price: Look for the offscreen span within this specific item
        price_span = item.find("span", class_="a-offscreen")
        price = price_span.get_text(strip=True) if price_span else "N/A"

        # Rating: Look for the star icon text
        rating_span = item.find("span", class_="a-icon-alt")
        rating = rating_span.get_text(strip=True) if rating_span else "N/A"

        all_products.append({
            "Title": title,
            "Price": price,
            "Rating": rating
        })
    except AttributeError:
        continue # Skip items that are missing data (like ads)

# Save the list of many items to CSV
df = pd.DataFrame(all_products)
df.to_csv("amazon_search_results.csv", index=False)

print(f"Done! Saved {len(all_products)} items to CSV.")
```
