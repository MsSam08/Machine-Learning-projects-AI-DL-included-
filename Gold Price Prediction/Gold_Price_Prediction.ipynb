### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

### Data Collection and Processing

```python
# loading the csv data to a Pandas DataFrame
gold_data = pd.read_csv('/content/gld_price_data.csv')
```

### Correlation:

Positive Correlation

Negative Correlation

```python
correlation = gold_data.drop('Date', axis=1).corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
plt.title('Correlation Matrix of Gold Price Data')
plt.show()
```

```python
# correlation values of GLD
print(correlation['GLD'])
```

```python
# checking the distribution of the GLD Price
sns.histplot(gold_data['GLD'], kde=True)
plt.show()
```

### Splitting the Features and Target

```python
X = gold_data.drop(['Date','GLD'],axis=1)
Y = gold_data['GLD']
```

```python
print(X)
```

```python
print(Y)
```

### Splitting into Training data and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

## Model Training: Random Forest Regressor

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
# training the model
regressor.fit(X_train,Y_train)
```

### Model Evaluation

```python
# prediction on Test Data
test_data_prediction = regressor.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared error : ", error_score)
```

### Compare the Actual Values and Predicted Values in a Plot

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color='blue', label = 'Actual Value')
plt.plot(test_data_prediction, color='green', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```

```python
r2 = regressor.score(X_test, Y_test)
print("R squared error:", r2)
```

### Conclusion

The gold price prediction model achieved an R-squared value of 0.9897, indicating that approximately 99% of the variation in gold prices is explained by the model. This demonstrates excellent predictive performance and confirms the effectiveness of the selected features and modeling approach. The results highlight the potential of machine learning techniques for accurate gold price forecasting.
